## Apply + lambda

```python
data.head()
```

```python
data['income_category'].unique()
```

Если клиента зарабатывает больше 80К, то добавляем 1000 к кредитному лимиту.  
Если клиент зарабатывает меньше 40К, то уменьшаем на 1000 его кредитный лимит.  
В противном случае оставляем всё как есть.  
Важное условие: кредитный лимит не должен превышать 30К

```python
def new_credit_limit(income_category, old_credit_limit):
    if income_category == 'Less than $40K':
        return max(0, old_credit_limit - 1000)
    elif income_category == '$80K - $120K' or income_category == '$120K +':
        return min(30000, old_credit_limit + 1000)
    else:
        return old_credit_limit
```

```python
%%timeit
data.apply(lambda x: new_credit_limit(x.income_category, x.avg_open_to_buy), axis=1)
```

```python
update_credit_limit = np.vectorize(new_credit_limit)
update_credit_limit(data.income_category, data.avg_open_to_buy)
```

```python
%%timeit
update_credit_limit(data.income_category, data.avg_open_to_buy)
```

```python
data['new_credit_limit'] = update_credit_limit(data.income_category, data.avg_open_to_buy)
```

## np.select

```python
data.head()
```

```python
data.income_category.unique()
```

```python
def get_level(data):
    return np.select(
        condlist=(
            data.income_category.isin(['Less than $40K']),
            data.income_category.isin(['$40K - $60K', '$60K - $80K']),
            data.income_category.isin(['$80K - $120K', '$120K +'])
        ),
        choicelist=('level_3', 'level_2', 'level_1'),
        default='other'
    )
```

```python
data.assign(level = lambda x: get_level(x))
```
