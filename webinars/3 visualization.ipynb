# Типы графиков

## 1. Изменения во времени

### 1.1 Line plot

Лучше использовать, когда хотите показать временную зависимость

```python
df = pd.read_csv('GlobalLandTemperaturesByCountry.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['dt'] = pd.to_datetime(df['dt'], format="%Y-%m-%d")
```

```python
df.info()
```

```python
df['dt'].min(), df['dt'].max()
```

```python
df_test = df.loc[df['dt'] > '2010-01-01 00:00:00'].copy()
```

`Line plot` отображает эволюцию одной или нескольких числовых переменных. Это один из наиболее распространенных типов диаграмм, поэтому его можно построить с использованием любой библиотеки python

```python
plt.plot(df_test.groupby('dt')['AverageTemperature'].mean(), marker='o')
plt.grid()
plt.xticks(rotation=45)
plt.show()
```

### 1.2 Bar plot

Столбчатая диаграмма показывает взаимосвязь между числовой и категориальной переменной.  
Каждая сущность категориальной переменной представлена в виде столбика.  
Размер полосы представляет ее числовое значение.

```python
df_test['month'] = df_test['dt'].apply(lambda x: x.month)
```

```python
mean_temp_group = df_test.groupby('month')['AverageTemperature'].mean()
mean_temp_group
```

```python
plt.bar(mean_temp_group.index, mean_temp_group)
```

```python
df_test['month_name'] = df_test['dt'].apply(lambda x: x.month_name())
df_test['month_name'].head()
```

```python
mean_temp_group = df_test.groupby('month_name')['AverageTemperature'].mean()\
    .reindex(['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November','December'])
mean_temp_group
```

```python
plt.bar(mean_temp_group.index, mean_temp_group)
plt.xticks(rotation=70);
```

### 1.3 Area plot

`Area plot` похож на `line plot`, за исключением того, что область между осью x и линией заполнена цветом или штриховкой.  
Он представляет эволюцию числовой переменной. 

Лучше использовать, когда вы хотите показать объем, а не тенденции.

### 2.1 Histogram

```python
data = pd.read_csv('StudentsPerformance.csv')
```

```python
data.columns = [c.replace(' ', '_').replace('/', '_') for c in data.columns]
data = data.rename(columns={'test_preparation_course':'is_prepared'})
```

```python
data.head()
```

```python
plt.hist(data['math_score'], bins=15, color='y', edgecolor='g') # rwidth=0.9 density=True
plt.grid()
plt.xlabel('math score')
plt.title('Histogram of math score')
pass
```

```python
np.linspace(0, 100, 16)
```

```python
plt.hist(data[['math_score', 'reading_score', 'writing_score']], stacked=True, alpha = 0.7, edgecolor='k', color=['blue', 'red', 'yellow'])
plt.grid()
plt.show()
```

```python
groups = sorted(data['race_ethnicity'].unique())

_, axs = plt.subplots(2, 3, figsize=(10, 8))
axs = axs.ravel()

for i, group in enumerate(groups):
    group = group
    ax = axs[i]
    
    ax.hist(data.loc[data['race_ethnicity']==group, 'math_score'], edgecolor='black')
    ax.grid()
    ax.set_title(group)
    
axs[-1].set_visible(False)
```

### 2.4 Box plot

**Кванти́ль** — значение, которое заданная случайная величина не превышает с фиксированной вероятностью.  
**Квартиль** - это тип квантиля, который делит количество точек данных на четыре части или четверти более или менее одинакового размера  
  
0,25 квантиль = 1 квартиль = 25 персентиль

Нижняя граница = Первый квартиль – 1.5 * Межквартильное расстояние,

Верхняя граница = Третий квартиль + 1.5 * Межквартильное расстояние,

Межквартильное расстояние (IQR) = 75 персентиль – 25 персентиль

```python
plt.boxplot(data['math_score'])
pass
```
