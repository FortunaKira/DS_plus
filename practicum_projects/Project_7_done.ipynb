## <a id="project-description">Описание проекта  </a>  

Интернет-магазин «В один клик» продаёт разные товары: для детей, для дома, мелкую бытовую технику, косметику и даже продукты. Отчёт магазина за прошлый период показал, что активность покупателей начала снижаться. Привлекать новых клиентов уже не так эффективно: о магазине и так знает большая часть целевой аудитории. Возможный выход — **удерживать активность постоянных клиентов**. Сделать это можно с помощью персонализированных предложений. «В один клик» — современная компания, поэтому её руководство не хочет принимать решения просто так — только на основе анализа данных и бизнес-моделирования.  

1. Нужно промаркировать уровень финансовой активности постоянных покупателей. В компании принято выделять два уровня активности: «снизилась», если клиент стал покупать меньше товаров, и «прежний уровень».
2. Нужно собрать данные по клиентам по следующим группам:
 - Признаки, которые описывают коммуникацию сотрудников компании с клиентом.
 - Признаки, которые описывают продуктовое поведение покупателя. Например, какие товары покупает и как часто.
 - Признаки, которые описывают покупательское поведение клиента. Например, сколько тратил в магазине.
 - Признаки, которые описывают поведение покупателя на сайте. Например, как много страниц просматривает и сколько времени проводит на сайте.


1. Нужно построить модель, которая предскажет вероятность снижения покупательской активности клиента в следующие три месяца.
2. В исследование нужно включить дополнительные данные финансового департамента о прибыльности клиента: какой доход каждый покупатель приносил компании за последние три месяца.
3. Используя данные модели и данные о прибыльности клиентов, нужно выделить сегменты покупателей и разработать для них персонализированные предложения.

## <a id="data-description">Описание данных</a>  


Данные для работы находятся в нескольких таблицах:  

**market_file.csv**  
Таблица, которая содержит данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении.  
- `id` — номер покупателя в корпоративной базе данных.
- `Покупательская активность` — рассчитанный класс покупательской активности (целевой признак): «снизилась» или «прежний уровень».
- `Тип сервиса` — уровень сервиса, например «премиум» и «стандарт».
- `Разрешить сообщать` — информация о том, можно ли присылать покупателю дополнительные предложения о товаре. Согласие на это даёт покупатель.
- `Маркет_актив_6_мес` — среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев. Это значение показывает, какое число рассылок, звонков, показов рекламы и прочего приходилось на клиента.
- `Маркет_актив_тек_мес` — количество маркетинговых коммуникаций в текущем месяце.
- `Длительность` — значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте.
- `Акционные_покупки` — среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев.
- `Популярная_категория` — самая популярная категория товаров у покупателя за последние 6 месяцев.
- `Средний_просмотр_категорий_за_визит` — показывает, сколько в среднем категорий покупатель просмотрел за визит в течение последнего месяца.
- `Неоплаченные_продукты_штук_квартал` — общее число неоплаченных товаров в корзине за последние 3 месяца.
- `Ошибка_сервиса` — число сбоев, которые коснулись покупателя во время посещения сайта.
- `Страниц_за_визит` — среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца.  

**market_money.csv**  
Таблица с данными о выручке, которую получает магазин с покупателя, то есть сколько покупатель всего потратил за период взаимодействия с сайтом.
- `id` — номер покупателя в корпоративной базе данных.
- `Период` — название периода, во время которого зафиксирована выручка. Например, 'текущий_месяц' или 'предыдущий_месяц'. *(поправка к базовому описанию: и пред_предыдущий_месяц)*
- `Выручка` — сумма выручки за период.  

**market_time.csv**  
Таблица с данными о времени (в минутах), которое покупатель провёл на сайте в течение периода.  
- `id` — номер покупателя в корпоративной базе данных.
- `Период` — название периода, во время которого зафиксировано общее время.
- `минут` — значение времени, проведённого на сайте, в минутах.

**money.csv**  
Таблица с данными о среднемесячной прибыли покупателя за последние 3 месяца: какую прибыль получает магазин от продаж каждому покупателю.  
- `id` — номер покупателя в корпоративной базе данных.
- `Прибыль` — значение прибыли.

- Клиенты с постоянной покупательской активностью реже оставляют неоплаченные продукты
- Ошибка сервиса, похоже, не влияет на покупательскую активность
- Чем больше страниц просматривает покупатель на сайте, тем ниже вероятность, что его покупательская активность снизится  
- Тип сервиса не влияет на покупательскую активность
- Разрешение на рассылку маркетинговых материалов не влияет на покупательскую активность

## <a id="df-union">Объединение df  </a>  

Создадим общий df  для дальнейшей работы, где `market_total = market_file + market_money + market_time`:

\* *данные о выручке и времени на сайте находятся в одном столбце для всех периодов, в итоговой таблице нужен отдельный столбец для каждого периода*

````python
market_total = market_file.copy() # создаём копию основного df

# создаём df-ы из market_money разделённые по периодам
market_pred_pred_period = (
    market_money[market_money['Период'] == 'пред_предыдущий_месяц']
    .rename(columns={'Выручка':'Выручка_пред_предыдущий_месяц'})
    [['id', 'Выручка_пред_предыдущий_месяц']]
)
market_pred_period = (
    market_money[market_money['Период'] == 'предыдущий_месяц']
    .rename(columns={'Выручка':'Выручка_предыдущий_месяц'})
    [['id', 'Выручка_предыдущий_месяц']]
)
market_period = (
    market_money[market_money['Период'] == 'текущий_месяц']
    .rename(columns={'Выручка':'Выручка_текущий_месяц'})
    [['id', 'Выручка_текущий_месяц']]
)
# добавляем информацию в основной df
market_total = market_total.merge(market_period, on='id', how='inner')
market_total = market_total.merge(market_pred_period, on='id', how='inner')
market_total = market_total.merge(market_pred_pred_period, on='id', how='inner')

# создаём df-ы из market_time разделённые по периодам
market_time_pred_period = (
    market_time[market_time['Период'] == 'предыдущий_месяц']
    .rename(columns={'Минут':'Минут_предыдущий_месяц'})
    [['id', 'Минут_предыдущий_месяц']]
)
market_time_period = (
    market_time[market_time['Период'] == 'текущий_месяц']
    .rename(columns={'Минут':'Минут_текущий_месяц'})
    [['id', 'Минут_текущий_месяц']]
)
# добавляем информацию в основной df
market_total = market_total.merge(market_time_pred_period, on='id', how='inner')
market_total = market_total.merge(market_time_period, on='id', how='inner')

market_total.head()
````

## <a id="using-pipelines">Использование пайплайнов  </a>  

Обучим несколько вариантов моделей (KNeighborsClassifier(), DecisionTreeClassifier(), LogisticRegression() и  SVC()) с несколькими вариантами кодировщиков и гиперпараметров. Для оценки качества моделей будем использовать ROC-AUC, т.к. необходимо оценить качество работы модели вне зависимости от выбранного порога вероятности и у нас дисбаланс целевого класса (воспользуемся стратификацией, но over/andersampling не будем применять, т.к. дисбаланс вызван не недосбором информации, а является нормой для бизнес-задачи).

<font color='blue'> Нужен код архитектуры Пайплайна. По тексту ошибки можно только сказать, что pipeline собран некорректно и некоторый его компоненты не "сериализуемые", то есть не воспринимаютсся как трансформер.\
Ещё может быть несоответствие версий библиотек. (sklearn)</font>

````python
X = market_total.drop(columns=['Покупательская_активность', 'id'], axis=1)
y = market_total['Покупательская_активность']

# разделяем данные на тренировочные и тестовые
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=TEST_SIZE,
    random_state=RANDOM_STATE,
    stratify=y
)

# создаём списки с названиями признаков
ohe_columns = ['Тип_сервиса', 'Разрешить_сообщать']
ord_columns = ['Популярная_категория']
num_columns = ['Маркет_актив_6_мес', 'Маркет_актив_тек_мес', 'Длительность',
               'Акционные_покупки', 'Средний_просмотр_категорий_за_визит',
               'Неоплаченные_продукты_штук_квартал','Ошибка_сервиса',
               'Страниц_за_визит', 'Выручка_текущий_месяц', 'Выручка_предыдущий_месяц', 
               'Выручка_пред_предыдущий_месяц', 'Минут_предыдущий_месяц', 'Минут_текущий_месяц']
# создаём пайплайн для подготовки признаков из списка ohe_columns
ohe_pipe = Pipeline(
    [
        (
            'simpleImputer_ohe', 
            SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        ),
        (
            'ohe', 
            OneHotEncoder(drop='first', handle_unknown='error', sparse=False)
        )
    ]
)
# создаём пайплайн для подготовки признаков из списка ord_columns: заполнение пропусков и Ordinal-кодирование
# SimpleImputer + OE
ord_pipe = Pipeline(
    [('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
     ('ord',  OrdinalEncoder(
                categories=[
                    ['Кухонная посуда', 'Мелкая бытовая техника и электроника', 
                 'Техника для красоты и здоровья', 'Косметика и аксесуары', 
                 'Домашний текстиль', 'Товары для детей'],
                ], 
                handle_unknown='use_encoded_value', unknown_value=np.nan
            )
        ),
     ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
    ]
)

# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [
        ('ohe', ohe_pipe, ohe_columns),
        ('ord', ord_pipe, ord_columns),
        ('num', MinMaxScaler(), num_columns)
    ], 
    remainder='passthrough'
)

# создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
    ('preprocessor', data_preprocessor),
    ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
])


param_grid = [
    {
        'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
        'models__max_depth': range(3, 7),
        'models__max_features': range(3, 7),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range(2, 10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [LogisticRegression(
            random_state=RANDOM_STATE,
            solver='liblinear',
            penalty='l1'
        )],
        'models__C': range(1, 10),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    },
    {
        'models': [SVC(kernel='sigmoid', probability=True)],
        'models__degree': range(3, 5),
        'models__C': range(1, 5),
        'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
    }
]

randomized_search = RandomizedSearchCV(
    pipe_final,
    param_grid,
    cv=10,
    scoring='roc_auc',
    random_state=RANDOM_STATE,
    n_jobs=-1
)

randomized_search.fit(X_train, y_train)
print('Результаты поиска:')
display(pd.DataFrame(randomized_search.cv_results_).sort_values(by='mean_test_score', ascending=False)[['param_models', 'param_preprocessor__num', 'mean_test_score', 'rank_test_score']])

print('Лучшая модель и её параметры:\n\n', randomized_search.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', randomized_search.best_score_)

y_test_pred = randomized_search.predict_proba(X_test)
print(f'Метрика ROC-AUC на тестовой выборке: {roc_auc_score(y_test, y_test_pred[:,1])}')

fpr, tpr, threshold = roc_curve(y_test, y_test_pred[:,1], pos_label='Снизилась')
plt.plot(fpr, tpr)
plt.title("График зависимости FPR от TPR")
plt.xlabel('False Positive Rate (FPR)')
plt.ylabel('True Positive Rate (TPR)')
plt.show()
````

Лучшая модель, которую удалось получить - LogisticRegression(C=6, penalty='l1', random_state=42, solver='liblinear'), с passthrough, где:  
- Метрика лучшей модели на тренировочной выборке: 0.8959340919867236
- Метрика ROC-AUC на тестовой выборке: 0.9155442300691664  

Какие выводы о важности признаков можем сделать:  
- Меньше всего на покупательскую активность клиента влияет тип сервиса, согласие на подписку, кол-во маркетинговых материалов в текущем месяце и давность первого прихода
- Больше всего влияет поведение на сайте: кол-во просмотренных страниц, категорий, время проведённое на сайте; и доля акционных покупок, неоплаченных товаров, полученных маркетинговых материалов за последние 6 месяцев

**При моделировании** можем попробовать **сократить список фичей**, убрать те, которые не оказывают значительного влияния на целевой признак, так модель станет проще, может повыситься точность.  

**Для бизнеса** важность этих признаков может стать ориентиром, **над чем имеет смысл поработать, чтобы улучшить покупательскую активность**.  
1. Если поведение на сайте имеет такую большую значимость, может, имеет смысл поработать над сайтом, чтобы он был ещё приятнее и удобнее пользователей, поработать над системой рекомендаций, чтобы пользователя "затягивало" при просмотре и он находил для себя больше интересующих товаров.  
2. Странно, что тип сервиса не влияет на покупатескую активность, может, премиум клиентам недостаточно бонусов и акции? Стоит подумать о стратегии программы лояльности. Сюда же можно и добавить поощрение клиентов. которые уже давно пользуются магазином, предлагать им бонусы на покупки или персональные скидки.
3. Большой дисбаланс между категориями товаров, возможно, кол-во артикулов просто на каждую категорию сильно отличается, а может, лучше всего проработаны только пара категорий. Стоит ли привлечь менеджера по закупкам, который проведёт анализ того, что имеется и составит план того, что действительно стоит закупить.

В первую очередь стоит обратить внимание на стратегию удержания среднеприбыльного сегмента, т.к. он самый многочисленный и, в общей сложности, является источником основной прибыли магазина. Предложения по работе с сегментом:  
 - "Средний покупатель" часто интересуется выгодными предложениями по программам лояльности. Стоит подумать о них, если их нет вообще, или модернизировать, если уже есть какие-то
 - Улучшить пользовательский опыт на сайте, т.к. имеем дело с  интернет-магазином, сайт - лицо компании. Он должен быть удобным, приятным глазу, интуитивно понятным, с лёгкой навигацией, с быстрой онлайн-поддержкой
 - Стоит оценить,какие цены у конкурентов на аналогичные товары, или может у них шире ассортимент
 - Можно попытаться получить обратную связь от клиента по результатам его покупок. Составить краткую форму обратной связи, которая могла бы помочь понять, где сервис или предложения оставляют желать лучшего
 - Точно стоит отдельно заинтересовать клиентов пользоваться премиум сервисом, т.к. сейчас тип сервиса хоть как-то взаимосвязан только с длительностью опыта взаимодействия клиента с магазином. Возможно, это просто статус, но который не даёт ничего существенно интересного, премиум клиенты не прибыльнее
